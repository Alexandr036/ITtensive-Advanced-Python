Чтобы пройти в следующий круг соревнований, футбольной команде нужно набрать хотя бы 4 очка в двух играх.<br> Если команда выигрывает, она получает 3 очка, в случае ничьей — 1 очко, если проигрывает — 0 очков.<br>  Найдите вероятность того, что команде удастся выйти в следующий круг соревнований. Считайте, что в каждой игре вероятности выигрыша и проигрыша одинаковы и равны 0,3.

Решение - Равновероятные события. P = m / N (где m - подходящие события, N - всего исходов)<br>
Всего 8 исходов  = 3-3, 3-1, 1-3, 1-1, 0-3, 3-0, 1-0, 0-1 <br> 
Проходных комбинаций  3 (сумма 2х игр 4 очка)<br> 

```python
N = 8
m = 3
p = m / N
print(f"Вероятность выхода команды в следующий круг соревнований равна {p}")
```

Таня заметила, что в казино “Подкинем” используют неправильную игральную кость (т.е. не у всех граней вероятности выпадения одинаковы). При этом она установила, что вероятность выпадения чётного числа равна 0,6; вероятность выпадения числа, делящегося на 3, равна 0,3; вероятность того, что выпадет 1 или 5, равна 0,22. Какова вероятность того, что на этой игральной кости выпадет число 3?

Решение - нужно узнать вероятность каждого числа и зная эту вероятность найти недостающую.<br>
Вер. четных = 0.6 т.е. вероятность 2 = 4 = 6 == 0.6 / 3 = 0.2<br>
Вер. 1 = 5 == 0.22 / 2 т.е. вероятность = 0.11<br>
Вероятность 3 = 1 - вер 1,2,4,5,6 == 1 - 0.6 - 0.22 == 0.18 <br>

```python
p_even = 0.6
p_2 = p_4 = p_6 = round(p_even / 3, 2)
p_1 = p_5 = 0.22/2
p_3 = 1 - p_1 - p_2 - p_4 - p_5 - p_6
print(f'Вероятность выпадения 3 = 1 - {p_1} - {p_2} - {p_4} - {p_5} - {p_6} = {round(p_3, 2)}')
```

Компания “Light” изготавливает лампочки. Вероятность того, что готовая лампочка неисправна, равна 0,04. Каждую лампочку дополнительно проверяет упаковщик. Вероятность того, что упаковщик обнаружит (и изымет) неисправную лампочку, равна 0,96. Вероятность того, что упаковщик по ошибке изымет исправную лампочку, равна 0,01. Найдите вероятность того, что случайно выбранная изготовленная лампочка будет изъята упаковщиком.

Решение. Найдем общее кол-во возможных лампочек для вынимания (неисправные + ошибочные ) = 0.04 + 0.01<br>
Найдем вероятность что случайно выбранная изготовленная лампочка будет изъята упаковщиком. Общее кол-во ламп для вынимания умножим на вероятность обнаружения. <br>
Произведение совместных событий P = P(A) * P(B)

```python
# произведение совместных событий P = P(A) * P(B)
p_a = 0.05 
p_b = 0.96
p = p_a * p_b
print('Вероятность того, что случайно выбранная изготовленная лампочка' '\n'   
        f'будет изъята упаковщиком равна {p}')
```

Чтобы поступить в университет на механико-математический факультет, абитуриент должен набрать на ЕГЭ не менее 75 баллов по каждому из трёх предметов – математика, русский язык и физика, и не менее 75 баллов за внутренний экзамен по математике.<br>
Чтобы поступить на факультет вычислительной математики и кибернетики, нужно набрать не менее 75 баллов за ЕГЭ по каждому из четырех предметов – математика, русский язык, физика и информатика, и не менее 60 баллов за тот же внутренний экзамен по математике, что и на механико-математический факультет.<br> 
Вероятность того, что абитуриент Cubert получит не менее 75 баллов за ЕГЭ по математике, равна 0,8, по русскому языку – 0,9, по физике – 0,85, по информатике – 0,7.<br> 
Вероятность того, что Cubert сдаст внутренний экзамен не менее, чем на 60 баллов равна 0,9. Вероятность того, что Cubert сдаст внутренний экзамен не менее, чем на 75 баллов равна 0,7.<br> 
Какова вероятность того, что Cubert’у хватит баллов хотя бы на один из двух упомянутых факультетов.

Вероятность поступить вычислительной математики и кибернетики  - сумма совместных событий:<br>
P(A) + P(B) - P(A) * P(B)


```python
mat = 0.8
rus = 0.9
fiz = 0.85
inform = 0.7
in_exam_60= 0.9
in_exam_75= 0.7


p_meh_mat = (mat*rus*fiz*in_exam_75)
print('Вероятность поступить на механико-математический факультет', round(p_meh_mat, 3))
p_mat_kiber = (mat*rus*fiz*inform*in_exam_60)
print('Вероятность поступить на факультет вычислительной математики и кибернетики', round(p_mat_kiber, 3))
```

Для поступления на оба факультета, Cubert'у нужно набрать по ЕГЭ все необходимые баллы, а также минимум по внутреннему экзамену 75, так как набор 75 баллов перекроет 60 баллов по одному из факультетов.

```python
# нахождение вероятности поступления на оба факультета
p_meh_mat_and_kiber = (mat * rus * fiz * inform * in_exam_75)
print(f"Вероятность, что набранных балов Cubert'у хватить на для поступления на оба факультета"
        '\n' f"равна {round(p_meh_mat_and_kiber, 3)}")
```

Для нахождения вероятности, что Cubert'у хватить хотя бы на 1 факультет, нужно сложить вероятности поступления на каждый факультет и вычесть вероятность поступления на оба факультета одновременно.

```python
# вероятность, что хотя бы на один факультет 
p_1 = (p_meh_mat + p_mat_kiber) - p_meh_mat_and_kiber
print(f"Вероятность, что Cubert поступит хотя бы на одни из факультетов равна {round(p_1, 3)}")
```
